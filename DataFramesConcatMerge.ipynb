# Merge

```python
df_merge1 = pd.read_csv('7-merge_data1.csv')
df_merge1
```

```python
df_merge2 = pd.read_csv('7-merge_data2.csv')
df_merge2
```

# Merge Outer Join

```python
df_merged_outer = pd.merge(df_merge1, df_merge2, on = 'Employee_ID', how = 'outer') # outer merge 
df_merged_outer
```

# Merge Left Join

```python
df_merged_left = pd.merge(df_merge1, df_merge2, on = 'Employee_ID', how = 'left') # left merge, keeping all records from df_merge1 (left dataframe)
df_merged_left
```

# Merge Right join

```python
df_merged_right = pd.merge(df_merge1, df_merge2, on = 'Employee_ID', how = 'right') # right merge, keeping all records from df_merge2 (right dataframe)
df_merged_right
```

# merge - inner join

```python
df_merged = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='inner') # merge dataframes on 'ID' column
df_merged
```
