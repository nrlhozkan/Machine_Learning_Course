```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Box Plot Exercise
# median, quartiles (25%, 50%, 75%), min max 1.5*IQR, outlier
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 40, 50, 55, 60, 100])
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y = data)
plt.title('Box Plot of Data')
plt.ylabel('Values')
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x = "class", y = "age", data = df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(x = "class", y = "fare", data = df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.tight_layout()
plt.show()
```
