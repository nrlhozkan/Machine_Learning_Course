```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('1-studyhours.csv')
df
```

```python
plt.scatter(df['Study Hours'], df['Exam Score'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
# independent and dependent features
# X -----> inputs
# y -----> output
```

```python
X = df[['Study Hours']]  # data frame olarak verilmesi beklenir.
y = df['Exam Score']
```

```python
# test - train split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
# standardize the data set
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# balanced feature values
# efficient gradient descent
# l1 and l2 
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
print("Coefficient: ", regression.coef_)
```

```python
print("Intercept:", regression.intercept_)
```

```python
# y = 76.91 + 16.18*X
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train), "r")
plt.show()
```

```python
# x =20, y = ?
```

```python
regression.predict([[20]]) #this is wrong
```

```python
x_transformed = scaler.transform([[20]])
regression.predict(x_transformed)
```

```python
# predict with test data
```

```python
y_pred_test = regression.predict(X_test)
```

```python
y_pred_test
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred_test, "r")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
print(f"MSE: {mse}, MAE: {mae}, RMSE: {rmse}")
```

```python
r2 = r2_score(y_test, y_pred_test)
print(f"R^2: {r2}")
```

```python
# adjusted r2 
r2_adjusted = 1 - (1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
