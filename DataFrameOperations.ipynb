# working with missing data

```python
weather_na_df = pd.read_excel('6-weatherna.xlsx')
weather_na_df
```

```python
weather_na_df.isna()
```

```python
weather_na_df['Paris'].count() # count of non-null values in 'Paris' column
```

```python
weather_na_df.dropna() # drop rows with any NaN values
```

```python
weather_na_df.dropna(axis = 1) # drop columns with any NaN values
```

```python
weather_na_df.drop('Paris', axis = 1) # drop 'Paris' column
```

```python
weather_na_df.fillna(20) # fill NaN values with 20
```

```python
weather_na_df.fillna(weather_na_df.mean()) # fill NaN values with column means
```

# Group By

```python
df = pd.read_csv('6-employee.csv')
df
```

```python
df.describe() # statistical summary of the DataFrame
```

```python
df[df['Experience']>6] # filter rows where 'Experience' is greater than 6
df[df['Experience']>6].count() # count of rows where 'Experience' is greater than 6
```

```python
df.head()
```

```python

df_grouped = df.groupby('Department') # group by 'Department' column
df_grouped.count() # count of rows in each group
```

```python
df_grouped.describe() # statistical summary of each group
```

```python
df_grouped['Salary'].mean() # mean salary for each department
```

```python
df_grouped_city = df.groupby(['City']) # group by 'City' column
df_grouped_city['Salary'].mean() # mean salary for each city
```

```python
df_grouped_city['Experience'].mean() # mean experience for each city
```
